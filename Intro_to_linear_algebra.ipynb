```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,100) # start, end, n points
```

Distance travelled by robber  d = 2.5t

```python
d_r = 2.5*t
```

Distance travelled by robber d = 3(t-5)

```python
d_s = 3*(t-5)
```

```python
fig , ax = plt.subplots()
plt.title("A bank robber caught")
plt.xlabel("time(in min)")
plt.ylabel("distance(in km)")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,d_r,c = 'green')
ax.plot(t,d_s,c = 'red')
plt.axvline(x = 30, color ='purple',linestyle = '--') # a vertical line
_ = plt.axhline(y = 75 , color = 'purple',linestyle = '--') # a horizontal line
```
